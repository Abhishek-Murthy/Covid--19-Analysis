# Pdf distribution of Total Cases

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15 ,6))
sns.distplot(data['Total Cases'])
plt.axvline(data['Total Cases'].mean(),color='black', label='Mean')
plt.title("Pdf distribution of Total Cases")
plt.legend(shadow=True,fontsize="larger")

skew = data['Total Cases'].skew()
kurt = data['Total Cases'].kurt()
print('Skewness:{}'.format(round(skew,2)))
print('Kurtosis:{}'.format(round(kurt,2)))
```

# Pdf distribution of Active Cases

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15 ,6))
sns.distplot(data['Active'])
plt.axvline(data['Active'].mean(),color='black', label='Mean')
plt.title("Pdf distribution of Active Cases")
plt.legend(shadow=True,fontsize="larger")

skew = data['Active'].skew()
kurt = data['Active'].kurt()
print('Skewness:{}'.format(round(skew,2)))
print('Kurtosis:{}'.format(round(kurt,2)))
```

# Pdf distribution of Deaths Cases

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15 ,6))
sns.distplot(data['Deaths'])
plt.axvline(data['Deaths'].mean(),color='black', label='Mean')
plt.title("Pdf distribution ofDeaths Cases")
plt.legend(shadow=True,fontsize="larger")

skew = data['Deaths'].skew()
kurt = data['Deaths'].kurt()
print('Skewness:{}'.format(round(skew,2)))
print('Kurtosis:{}'.format(round(kurt,2)))
```
